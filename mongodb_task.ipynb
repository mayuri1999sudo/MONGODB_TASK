```python
id=[]
name=[]
examscore=[]
quizscore=[]
homeworkscore=[]

for i in range(len(student)):
  id.append(student[i]['_id'])
  name.append(student[i]['name'])
  examscore.append(student[i]['scores'][0]['score'])
  quizscore.append(student[i]['scores'][1]['score'])
  homeworkscore.append(student[i]['scores'][2]['score'])
```

```python
d = zip(id,name,examscore,quizscore,homeworkscore)
df=pd.DataFrame(d, columns=['_id','name','exam_score', 'quiz_score', 'homework_score'])
df
```

```python
#Exporting data to mongodb
  
data = df.to_dict("records")
record1.insert_many(data)
```

```python
#1.Find the student name who scored maximum scores in all (exam, quiz and homework)?
score0=record1.find_one(sort=[('exam_score',-1)])
print("\nMaximum marks scored by student in exam is {}\n".format(score0['name']))
score1=record1.find_one(sort=[('quiz_score',-1)])
print("\nMaximum marks scored by student in quiz is {}\n".format(score1['name']))
score2=record1.find_one(sort=[('homework_score',-1)])
print("\nMaximum marks scored by student in homework is {}\n".format(score2['name']))
```

```python
#2.Find students who scored below average in the exam and pass mark is 40%?
exam_avg=df['exam_score'].mean()
exam_avg
passmark=40

for i in record1.find({"$and":[{"exam_score":{"$lt":exam_avg}},{"exam_score":{"$gt":passmark}}]},{'_id':1,"name":1}):
  print(i)
```

```python
#3.Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories
x = record1.aggregate(
[{
    "$project" :
    {
        "name":1,
        "_id" : 0,
        "result":{
            "$cond":
                    {"if": {"$and" : [{"$gte" : ["$exam_score", 40]}, {"$gte" : ["$quiz_score", 40]}, {"$gte" : ["$homework_score", 40]}]},
                    "then" :"pass",
                    "else":"fail"
                 }
                }
    }
}])
for i in x:
    print(i)
```
